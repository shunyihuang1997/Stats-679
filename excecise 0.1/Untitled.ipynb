```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(pd.read_csv('watches.csv',index_col = [0]))
df['Full price'] = df['New price']/((100-df['Discount'])/100)
df.head()
```

### Hypothesis: tommy hilfiger has the lowest mean full-price among these brands

```python
brand_mean_full_price = pd.DataFrame(df.groupby(['Name'])['Full price'].mean()).reset_index()
brand_mean_discount_price = pd.DataFrame(df.groupby(['Name'])['New price'].mean()).reset_index()
```

```python
#plt.title('Brand mean full-price')
brand_mean_full_price.sort_values('Full price', ascending=False).plot(x = 'Name',kind = 'bar', rot = 45, title = 'Brand mean full-price')
plt.tight_layout
```

### Hypothesis: timex has the lowest mean discount-price among these brands

```python
brand_mean_discount_price.sort_values('New price', ascending=False).plot(x = 'Name',kind = 'bar', rot = 45, title = 'Brand mean Discount price')
plt.tight_layout
```
